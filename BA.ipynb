````python
from pathlib import Path

path = Path("/content/drive/MyDrive/config.yml")

#@markdown #yml ç”¢ç”Ÿå™¨
#@markdown 1. è¼¸å…¥ä½ çš„å·´å“ˆå¸³è™ŸåŠå¯†ç¢¼ã€‚
#@markdown 2. å¦‚æœæœ‰é–‹å•Ÿå…©æ­¥é©Ÿé©—è­‰ï¼Œè¨˜å¾—è¼¸å…¥å…©æ­¥é©Ÿé©—è­‰ç¨®å­ç¢¼ï¼ˆä¸æ˜¯é‚£å…­ä½æ•¸å­—ï¼Œæ‡‰è©²æ˜¯ä¸€ä¸²å¤§å¯«è‹±æ–‡ï¼‰ã€‚

#@markdown ##ç™»å…¥è³‡è¨Š
username = "" #@param {type: 'string'}
password = "" #@param {type: 'string'}
twofa = "" #@param {type: 'string'}
#@markdown ---
#@markdown ##æŠ½çå˜—è©¦æ¬¡æ•¸
lottery = 20 #@param {type: 'number'}
#@markdown ---
#@markdown ##å¯„å ±å‘Šçš„éƒ¨åˆ†
#@markdown >å¦‚æœè¦å•Ÿç”¨ **Telegram Report** è«‹ä¾ç…§ä»¥ä¸‹æ­¥é©Ÿï¼š
#@markdown 1. é»æ“Šã€Œ [æ©Ÿå™¨äºº](https://t.me/automia_bot) ã€æŒ‰éˆ•
#@markdown 2. é–‹å•Ÿ Telegram æ‡‰ç”¨ç¨‹å¼ä¸¦èˆ‡æ©Ÿå™¨äººã€Œ **Automia** ã€é–‹å§‹å°è©±
#@markdown 3. å‚³é€ã€Œ**ID**ã€çµ¦æ©Ÿå™¨äººã€Œ **Automia** ã€
#@markdown 4. è¤‡è£½åƒã€Œ **1010101010** ã€çš„é€šé“ ID åˆ°å°æ‡‰çš„ç”¢ç”Ÿå™¨æ¬„ä½
telegram = "" #@param {type: 'string'}
#@markdown >å¦‚æœè¦å•Ÿç”¨ **Discord Report** è«‹ä¾ç…§ä»¥ä¸‹æ­¥é©Ÿï¼š
#@markdown 1. å»ºç«‹ä¸€å€‹è‡ªå·±çš„ **Discord** ä¼ºæœå™¨ï¼Œåœ¨ä½ æƒ³è¦ç™¼é€çš„æ–‡å­—é »é“æŒ‰ **é½’è¼ª** è¨­å®š
#@markdown 2. é»æ“Š æ•´åˆ -> **Webhook -> æ–° Webhook**
#@markdown 3. é»æ“Š è¤‡è£½ **Webhook ç¶²å€** ä¸¦è²¼åˆ°åˆ°å°æ‡‰çš„ç”¢ç”Ÿå™¨æ¬„ä½
discord = "" #@param {type: 'string'}
#@markdown >å¦‚æœè¦å•Ÿç”¨ **Line Notify Report** è«‹ä¾ç…§ä»¥ä¸‹æ­¥é©Ÿï¼š
#@markdown 1. é€²å…¥ [Line Notify](https://notify-bot.line.me/zh_TW/)
#@markdown 2. é»æ“Šå³ä¸Šæ–¹ç™»å…¥ Line å¸³è™Ÿ
#@markdown 3. é»æ“Šå³ä¸Šæ–¹ **è‡ªå·±çš„ ID -> å€‹äººé é¢ -> ç™¼è¡Œæ¬Šæ–**
#@markdown 4. å¡«å…¥åç¨±(æœƒé¡¯ç¤ºåœ¨è¨Šæ¯ä¸Š)ï¼Œä¸¦é¸æ“‡è¦æ¥æ”¶è¨Šæ¯çš„èŠå¤©å®¤æˆ–ç¾¤çµ„ï¼Œå»ºè­°é¸æ“‡ **é€é1å°1èŠå¤©æ¥æ”¶LINE Notifyçš„é€šçŸ¥**ï¼Œè‹¥é¸æ“‡ç¾¤çµ„çš„è©±ï¼Œéœ€å°‡ Line Notify å®˜æ–¹å¸³è™ŸåŠ å…¥ç¾¤çµ„å¾Œæ‰æ”¶çš„åˆ°è¨Šæ¯
#@markdown 5. æŒ‰ä¸‹ç™¼è¡Œ
#@markdown 6. é»æ“Š **è¤‡è£½**ï¼Œä¸¦å°‡æ¬Šæ–ä»£ç¢¼è²¼åˆ°å°æ‡‰çš„ç”¢ç”Ÿå™¨æ¬„ä½
line_notify = "" #@param {type: 'string'}
# æœ‰
if not path.exists():
  with open("/content/drive/MyDrive/config.yml", "w") as f:
    f.write("shared:\n")
    f.write("  shared_key: shared_val\n")
    f.write("modules:\n")
    f.write("  utils:\n")
    f.write("  report:\n")
    f.write("  ad_handler:\n")
    f.write("  login:\n")
    f.write("    username: \"{}\"\n".format(username))
    f.write("    password: \"{}\"\n".format(password))
    f.write("    twofa: \"{}\"\n".format(twofa))
    f.write("  sign:\n")
    f.write("  guild:\n")
    f.write("  answer:\n")
    f.write("  lottery:\n")
    f.write("    max_attempts: {}\n".format(str(lottery)))
    f.write("    max_parallel: 5\n")
    f.write("  del_mail:\n")
    f.write("    match:\n")
    f.write("      - title: \"ã€å‹‡è€…ç¦åˆ©ç¤¾ã€‘æˆåŠŸç²å¾—æŠ½çè³‡æ ¼é€šçŸ¥ä¿¡\"\n")
    f.write("  logout:\n")
    f.write("  telegram:\n")
    f.write("    channel: \"{}\"\n".format(telegram))
    f.write("  discord:\n")
    f.write("    webhook: \"{}\"\n".format(discord))
    f.write("  line_notify:\n")
    f.write("    token: \"{}\"\n".format(line_notify))
    f.write("  issue:\n")
    f.write("browser:\n")
    f.write("  type: firefox\n")
    f.write("  headless: true\n")
    f.write("  firefoxUserPrefs:\n")
    f.write("    dom.webaudio.enabled: false\n")
    f.write("    media.volume_scale: 0\n")
    f.write("    media.default_volume: 0\n")
    f.write("# executablePath: /path/to/browser\n")
else:
  # å¯«å…¥æª”æ¡ˆ
  with open("/content/drive/MyDrive/config.yml", 'r+') as f:
    content = f.readlines()
    for i, line in enumerate(content):

      if "username:" in line:
        line = "    username: \"{}\"\n".format(username)
        content[i] = line
      if "password:" in line:
        line = "    password: \"{}\"\n".format(password)
        content[i] = line
      if "twofa:" in line:
        line = "    twofa: \"{}\"\n".format(twofa)
        content[i] = line
      if "max_attempts:" in line:
        line = "    max_attempts: {}\n".format(str(lottery))
        content[i] = line
      if "channel:" in line:
        line = "    channel: \"{}\"\n".format(telegram)
        content[i] = line
      if "webhook:" in line:
        line = "    webhook: \"{}\"\n".format(discord)
        content[i] = line
      if "token:" in line:
        line = "    token: \"{}\"\n".format(line_notify)
        content[i] = line

      f.seek(0)
      f.writelines(content)
````
