````python
from pathlib import Path

path = Path("/content/drive/MyDrive/config.yml")

#@markdown #yml 產生器
#@markdown 1. 輸入你的巴哈帳號及密碼。
#@markdown 2. 如果有開啟兩步驟驗證，記得輸入兩步驟驗證種子碼（不是那六位數字，應該是一串大寫英文）。

#@markdown ##登入資訊
username = "" #@param {type: 'string'}
password = "" #@param {type: 'string'}
twofa = "" #@param {type: 'string'}
#@markdown ---
#@markdown ##抽獎嘗試次數
lottery = 20 #@param {type: 'number'}
#@markdown ---
#@markdown ##寄報告的部分
#@markdown >如果要啟用 **Telegram Report** 請依照以下步驟：
#@markdown 1. 點擊「 [機器人](https://t.me/automia_bot) 」按鈕
#@markdown 2. 開啟 Telegram 應用程式並與機器人「 **Automia** 」開始對話
#@markdown 3. 傳送「**ID**」給機器人「 **Automia** 」
#@markdown 4. 複製像「 **1010101010** 」的通道 ID 到對應的產生器欄位
telegram = "" #@param {type: 'string'}
#@markdown >如果要啟用 **Discord Report** 請依照以下步驟：
#@markdown 1. 建立一個自己的 **Discord** 伺服器，在你想要發送的文字頻道按 **齒輪** 設定
#@markdown 2. 點擊 整合 -> **Webhook -> 新 Webhook**
#@markdown 3. 點擊 複製 **Webhook 網址** 並貼到到對應的產生器欄位
discord = "" #@param {type: 'string'}
#@markdown >如果要啟用 **Line Notify Report** 請依照以下步驟：
#@markdown 1. 進入 [Line Notify](https://notify-bot.line.me/zh_TW/)
#@markdown 2. 點擊右上方登入 Line 帳號
#@markdown 3. 點擊右上方 **自己的 ID -> 個人頁面 -> 發行權杖**
#@markdown 4. 填入名稱(會顯示在訊息上)，並選擇要接收訊息的聊天室或群組，建議選擇 **透過1對1聊天接收LINE Notify的通知**，若選擇群組的話，需將 Line Notify 官方帳號加入群組後才收的到訊息
#@markdown 5. 按下發行
#@markdown 6. 點擊 **複製**，並將權杖代碼貼到對應的產生器欄位
line_notify = "" #@param {type: 'string'}
# 有
if not path.exists():
  with open("/content/drive/MyDrive/config.yml", "w") as f:
    f.write("shared:\n")
    f.write("  shared_key: shared_val\n")
    f.write("modules:\n")
    f.write("  utils:\n")
    f.write("  report:\n")
    f.write("  ad_handler:\n")
    f.write("  login:\n")
    f.write("    username: \"{}\"\n".format(username))
    f.write("    password: \"{}\"\n".format(password))
    f.write("    twofa: \"{}\"\n".format(twofa))
    f.write("  sign:\n")
    f.write("  guild:\n")
    f.write("  answer:\n")
    f.write("  lottery:\n")
    f.write("    max_attempts: {}\n".format(str(lottery)))
    f.write("    max_parallel: 5\n")
    f.write("  del_mail:\n")
    f.write("    match:\n")
    f.write("      - title: \"【勇者福利社】成功獲得抽獎資格通知信\"\n")
    f.write("  logout:\n")
    f.write("  telegram:\n")
    f.write("    channel: \"{}\"\n".format(telegram))
    f.write("  discord:\n")
    f.write("    webhook: \"{}\"\n".format(discord))
    f.write("  line_notify:\n")
    f.write("    token: \"{}\"\n".format(line_notify))
    f.write("  issue:\n")
    f.write("browser:\n")
    f.write("  type: firefox\n")
    f.write("  headless: true\n")
    f.write("  firefoxUserPrefs:\n")
    f.write("    dom.webaudio.enabled: false\n")
    f.write("    media.volume_scale: 0\n")
    f.write("    media.default_volume: 0\n")
    f.write("# executablePath: /path/to/browser\n")
else:
  # 寫入檔案
  with open("/content/drive/MyDrive/config.yml", 'r+') as f:
    content = f.readlines()
    for i, line in enumerate(content):

      if "username:" in line:
        line = "    username: \"{}\"\n".format(username)
        content[i] = line
      if "password:" in line:
        line = "    password: \"{}\"\n".format(password)
        content[i] = line
      if "twofa:" in line:
        line = "    twofa: \"{}\"\n".format(twofa)
        content[i] = line
      if "max_attempts:" in line:
        line = "    max_attempts: {}\n".format(str(lottery))
        content[i] = line
      if "channel:" in line:
        line = "    channel: \"{}\"\n".format(telegram)
        content[i] = line
      if "webhook:" in line:
        line = "    webhook: \"{}\"\n".format(discord)
        content[i] = line
      if "token:" in line:
        line = "    token: \"{}\"\n".format(line_notify)
        content[i] = line

      f.seek(0)
      f.writelines(content)
````

```
    shared_key: shared_val

# 使用的模組及其參數，可以是內建的模組或是自製模組的路徑

    utils:
    report:
    ad_handler:

    # 主要功能模組
    login:
        username: "a0123456789"
        password: "password"
        twofa: "AAAABBBBCCCCDDDD"
    sign:
    guild:
    answer:
    lottery:
        max_attempts: 100
        max_parallel: 5
    del_mail:
        match:
          - title: "【勇者福利社】成功獲得抽獎資格通知信"
    logout:

    # 以下是寄報告的部分
    telegram:
        channel: telegram_channel_id
    discord:
        webhook: "https://discord.com/api/webhooks/..."
    line_notify:
        token: line_token
    issue:

    # 自製模組（範例）
    # /path/to/module:
    #     param1: value1
    #     param2: value2

# 瀏覽器的設定
browser:
    type: firefox
    headless: true
    firefoxUserPrefs:
        dom.webaudio.enabled: false
        media.volume_scale: 0
        media.default_volume: 0
    # executablePath: /path/to/browser

    # 更多設定請查閱：https://playwright.dev/docs/api/class-browsertype#browser-type-launch
```
